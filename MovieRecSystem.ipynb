```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import os 
```

```python
print(os.getcwd())
os.chdir('C:\\Users\\Madhura\\Downloads\\')
#loading the data frpm the csv file
movie_data= pd.read_csv('movies.csv')
```

```python
#feature selection
features=['genres','keywords','original_language','title','popularity','cast','vote_count','director']
print(features)
```

```python
#replacing missing values to handle textaul dat efficiently
for feature in features:
    movie_data[feature] = movie_data[feature].fillna('')

allfeatures_data = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['original_language']+' '+movie_data['title']+' '+movie_data['cast']+' '+movie_data['director']
display(allfeatures_data)
```

```python
vector= TfidfVectorizer()
feature_vector=vector.fit_transform(allfeatures_data)
print(feature_vector)
```

```python
#similarity confidence by comparing one movie with every other movie in the data set
similarity= cosine_similarity(feature_vector)
print  (similarity )
print(similarity.shape)
```
